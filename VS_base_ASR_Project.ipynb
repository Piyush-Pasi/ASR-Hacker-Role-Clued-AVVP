```python
def intersection(a,b):
  a_s, a_e = a[0], a[1]
  b_s, b_e = b[0], b[1]
  if a_e <= b_s: return 0.0
  if a_s >= b_e: return 0.0
  m = max(a_s, b_s)
  n = min(a_e, b_e)
  u_m = min(a_s,b_s)
  u_n = max(a_e,b_e)
  return (n-m)/(u_n-u_m)
```

```python
def overlap(gt,vad_pred, round_num=False):
  gt_s = gt[0]
  gt_e = gt[1]

  org_i, s1_i, s2_i = 0.0, 0.0, 0.0
  if vad_pred[0].shape[-1]:
    org_s, org_e = vad_pred[0][0,0],vad_pred[0][0,1]
    if round_num:
      org_s, org_e = torch.round(org_s).item(), torch.round(org_e).item()
    org_i = intersection((gt_s,gt_e),(org_s,org_e))
  if vad_pred[1].shape[-1]:
    org_s, org_e = vad_pred[1][0,0],vad_pred[1][0,1]
    if round_num:
      org_s, org_e = torch.round(org_s).item(), torch.round(org_e).item()
    s1_i = intersection((gt_s,gt_e),(org_s,org_e))
  if vad_pred[2].shape[-1]:
    org_s, org_e = vad_pred[2][0,0],vad_pred[2][0,1]
    if round_num:
      org_s, org_e = torch.round(org_s).item(), torch.round(org_e).item()
    s2_i = intersection((gt_s,gt_e),(org_s,org_e))
  if org_i == 0.0 : org_i = s1_i
  return org_i, s1_i, s2_i
  


```

## MIoU


We used validation set data, it has timestamps for speech. We got timestamps for same file from Source-Separator+VAD and mIoU is shown below.

mean IoU for original final (direct VAD, no source-separator), source separated for speaker,  source separated for background

```python
mean_org, mean_s1, mean_s2 = 0.0, 0.0, 0.0
n = 0
for f in all_speech_files.keys():
  if f in all_boundaries.keys():
    a,b,c = overlap(all_speech_files[f],all_boundaries[f])
    mean_org += a
    mean_s1 += b
    mean_s2 += c
    n += 1

print(mean_org/n, mean_s1/n, mean_s2/n)
```
